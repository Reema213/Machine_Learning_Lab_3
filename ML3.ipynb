```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
# Load Dataset

df = pd.read_csv(r"C:\Users\remaq\Downloads\archive\IRIS.csv")

# Display first 5 rows
df.head()
```

```python
# viewing the data types of columns
df.dtypes
```

```python
print("Dataset Shape (rows, columns):", df.shape,"\n")
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
#convert species to categorical
df['species'] = df['species'].astype('category')

# Check the new types
print(df.dtypes)
```

```python
# Verify the change
print("Updated Data Types:")
print(df.dtypes)
```

```python
# Check for missing values
print("Missing Values:")
print(df.isna().sum())
```

```python
# Check for duplicate rows
duplicates = df.duplicated()
print("Number of duplicate rows:", duplicates.sum())
if duplicates.sum() > 0:
    print("Duplicate rows:")
    print(df[duplicates])
else:
    print("No duplicate rows found.")
```

```python
# Descriptive Statistics Summary
print("Descriptive Statistics for Numerical Features:")
df.describe()
```

```python
# Summary for categorical features
print("Species Distribution:")
print(df['species'].value_counts())
print("\nSpecies Distribution (percentages):")
print(df['species'].value_counts(normalize=True) * 100)
```

# UNIVARIATE ANALYSIS

```python
# Distribution of Sepal Length
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='sepal_length', bins=20, kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')

plt.subplot(1, 2, 2)
sns.boxplot(data=df, y='sepal_length')
plt.title('Box Plot of Sepal Length')
plt.ylabel('Sepal Length (cm)')
plt.tight_layout()
plt.show()
```

```python
# Distribution of Petal Length
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='petal_length', bins=20, kde=True)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')

plt.subplot(1, 2, 2)
sns.boxplot(data=df, y='petal_length')
plt.title('Box Plot of Petal Length')
plt.ylabel('Petal Length (cm)')
plt.tight_layout()
plt.show()
```

```python
# Distribution of Petal Width
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='petal_width', bins=20, kde=True)
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')

plt.subplot(1, 2, 2)
sns.boxplot(data=df, y='petal_width')
plt.title('Box Plot of Petal Width')
plt.ylabel('Petal Width (cm)')
plt.tight_layout()
plt.show()
```

# BIVARIATE ANALYSIS

```python
# Pairplot to see relationships between all features
sns.pairplot(df, hue='species', diag_kind='kde')
plt.suptitle('Pairplot of Iris Features by Species', y=1.02)
plt.show()
```

```python
# Sepal Length vs Sepal Width by Species
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', s=100)
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# Petal Length vs Petal Width by Species
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', s=100)
plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# Box plots comparing features across species
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
# Sepal Length by Species
sns.boxplot(data=df, x='species', y='sepal_length', ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length by Species')
axes[0, 0].set_xlabel('Species')
axes[0, 0].set_ylabel('Sepal Length (cm)')

# Sepal Width by Species
sns.boxplot(data=df, x='species', y='sepal_width', ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width by Species')
axes[0, 1].set_xlabel('Species')
axes[0, 1].set_ylabel('Sepal Width (cm)')

# Petal Length by Species
sns.boxplot(data=df, x='species', y='petal_length', ax=axes[1, 0])
axes[1, 0].set_title('Petal Length by Species')
axes[1, 0].set_xlabel('Species')
axes[1, 0].set_ylabel('Petal Length (cm)')

# Petal Width by Species
sns.boxplot(data=df, x='species', y='petal_width', ax=axes[1, 1])
axes[1, 1].set_title('Petal Width by Species')
axes[1, 1].set_xlabel('Species')
axes[1, 1].set_ylabel('Petal Width (cm)')

plt.tight_layout()
plt.show()
```

# CORRELATION ANALYSIS

```python
# Correlation matrix for numerical features
correlation_matrix = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Iris Features')
plt.tight_layout()
plt.show()

print("Correlation Interpretation:")
print("- Values close to 1: Strong positive correlation")
print("- Values close to -1: Strong negative correlation")
print("- Values close to 0: Weak or no correlation")
print("\nKey Observations:")
print("1. Petal length and petal width are highly positively correlated (0.96)")
print("2. Petal features have strong positive correlation with sepal length")
print("3. Sepal width shows weak correlation with other features")
```
